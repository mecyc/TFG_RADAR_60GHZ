```python
# Cada instancia es independiente
car_np = np.concatenate(list(map(lambda x: np.array(x), cartones)))
pla_np = np.concatenate(list(map(lambda x: np.array(x), plasticos)))
cri_np = np.concatenate(list(map(lambda x: np.array(x), cristales)))

cartondf = pd.DataFrame(car_np)
plasticodf = pd.DataFrame(pla_np)
cristaldf = pd.DataFrame(cri_np)
```

```python
# Se agrupan las instancias 

def byBatch(df, batch_size=500):
    pap = []
    for batch_number, batch_df in df.groupby(np.arange(len(df)) // batch_size):
        pap.append(batch_df.mean(axis=0))
    return pd.DataFrame(pap)

bsize = 100
cartondf = byBatch(cartondf, bsize)
plasticodf = byBatch(plasticodf, bsize)
cristaldf = byBatch(cristaldf, bsize)
```
